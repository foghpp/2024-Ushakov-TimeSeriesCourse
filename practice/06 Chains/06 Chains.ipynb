## **Часть 2. Визуализация цепочек**

Выполните визуализацию найденных в прошлом задании цепочек: постройте график временного ряда, на котором выделены звенья цепочек. Также постройте график, на котором изображены только сами звенья.

Визуализация цепочек с графиком временного ряда:

```python
fig, ax = plt.subplots(figsize=(16, 2))

ax.plot(ts)
for i in range(unanchored_chain.shape[0]):
    x = np.arange(unanchored_chain[i], unanchored_chain[i]+m)
    y = ts[x]
    ax.plot(x, y, color='red')
```

Визуализация цепочек без графика временного ряда:

```python
fig, ax = plt.subplots(figsize=(16, 2))

for i in range(unanchored_chain.shape[0]):
    x = np.arange(unanchored_chain[i], unanchored_chain[i]+m)
    y = ts[x]
    ax.plot(x, y, color='red')
```

❓ Проанализируйте и изложите содержательный смысл полученных результатов.

Как видно из визуализации звеньев цепочек временного ряда, их форма постепенно
изменяется - они сужаются или расширяются, за счет чего при равном скользящем
окне приобретают или теряют (соответственно) новые детали. Также звенья
изменяют свою форму на верхней их части, где в начале явно видно два пика и
один минимум, а к концу они начинают сглаживаться.
